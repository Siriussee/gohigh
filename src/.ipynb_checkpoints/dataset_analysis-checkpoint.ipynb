```python
import pandas as pd
```

## Clean Dataset

```python
data = pd.read_csv('contract_dataset2.csv')
contracts_df = data[~data['compiler_version'].str.contains('vyper', regex=False)]
contracts_df = contracts_df.reset_index().drop(columns=['index'])
contracts_df.info()
```

```python
def clean_code(code):
    code = code.strip("'")
    code = code.strip('"')
    code = code.replace(r'\r\n', '\n')
    code = code.replace(r'\n', '\n')
    code = code.replace(r'\r', '')
    code = code.replace(r'\t', '\t')
    code = code.replace(r"\'", '')
    code = code.replace(r"''", "'") # newly added
    return code

contracts_df['code'] = contracts_df['code'].apply(clean_code) 
```

```python
mask = pd.Series([True if 'abi.encode(' in x else False for x in contracts_df['code']].unique())
```

```python
def extract_verson_number(string):
    return string.strip('v').split('+')[0].split('-')[0]

contracts_df['compiler_version'] = contracts_df['compiler_version'].apply(lambda x: extract_verson_number(x))
```

```python
count_arb_call, count_const_call = 0, 0
count_send, count_assembly = 0, 0
count_none = 0

for contract in contracts:
    if contract.find(".call(") != -1:
        if contract.find(".call(bytes4") != -1 or contract.find(".call(abi.encode") != -1:
            count_const_call += 1
        else:
            count_arb_call += 1    
    elif contract.find(".send(") != -1:
        count_send += 1
    elif contract.find(":= call(")!= -1:
        count_assembly += 1   
    else:
        count_none += 1

#contract_types = [get_type2(c) for c in contracts]

print("count_arb_call: {}, count_const_call: {}, count_send: {}, count_assembly: {}, count_none: {}".format(
    count_arb_call, count_const_call, count_send, count_assembly, count_none))
```

```python
def get_type(code):
    pattern = ''
    if '.send(' in code:
        pattern += 'SE'
    if '.call(' in code:
        if '.call(bytes4(' in code or '.call(abi.encode' in code:
            pattern += 'CS'
        else:
            pattern += 'AB'
    if ':= call(' in code:
        pattern += 'IA'
    return pattern

from collections import Counter

contract_types = [get_type(c) for c in contracts]
print(Counter(contract_types))


ucontract_types = [get_type(c) for c in unique_contracts]
print(Counter(ucontract_types))
```

```python
with open('compile_send.sh', 'w') as f:
    for _, contract in send_contract_dd.iterrows():
        addr = contract['address']
        fn = f'../contract_send/{addr}.sol'
        out_fn = f'../contract_send_output/{addr}.err'
        version = contract['compiler_version']
        switch_commmand = f'solc-select use {version}'
        compile_command = f'solc {fn} > /dev/null 2> {out_fn}'
        f.write(switch_commmand + '\n')
        f.write(compile_command + '\n')
        
#cd sourcecode/ & !sh compile_send.sh
```

## Send Contract Stats

```python
import os

fns = [fn for _, _, fn in os.walk('../contract_send_output')]
addrs, passes = [], []

for fn in fns[0]:
    with open(os.path.join('../contract_send_output', fn)) as f:
        err_msg = f.read()
        addrs.append(fn.split('.')[0])
        passes.append(False if 'Error' in err_msg else True)
       
```

```python
compiled_contract = {'address': addrs, 'pass': passes}
compiled_contract_df = pd.DataFrame.from_dict(compiled_contract)
compiled_contract_df
```

```python
print('unique corner cases:', 
      compiled_contract_df[compiled_contract_df['pass']==True].count() / compiled_contract_df.count())
```

```python
send_contract_compiled = pd.merge(compiled_contract_df, send_contract_dd, how="left", on=["address"])
```

```python
import hashlib
def hash_func(code):
    return hashlib.sha224(bytes(code, 'utf-8')).hexdigest()

send_contract['hash_code'] = send_contract['code'].apply(lambda x: hash_func(x))
send_contract_compiled['hash_code'] = send_contract_compiled['code'].apply(lambda x: hash_func(x))
```

```python
# I'd better not report the overall result.

send_contract_compiled_all = pd.merge(send_contract, 
                                      send_contract_compiled[['pass', 'hash_code']], 
                                      how="left", on=["hash_code"])
print('unique corner cases:', 
      send_contract_compiled_all[send_contract_compiled_all['pass']==True].count() / send_contract_compiled_all.count())
```
